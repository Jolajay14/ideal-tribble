```
#read FaultDataset into spark dataframe

FaultDataset_DF = spark.read.csv("/FileStore/tables/FaultDataset.csv",
                                 header = "true",
                                 inferSchema="true")
```

```
# look at the schema of the DataFrame
FaultDataset_DF.printSchema()
```

```
# use display to view faultDataset dataframe
FaultDataset_DF.display()
```

```
# Check for missing values in each column in the fault Dataset DataFrame
from pyspark.sql.functions import sum

missing_values_count = FaultDataset_DF.agg(*(sum(FaultDataset_DF[col].isNull().cast("int")).alias(col) for col in FaultDataset_DF.columns))

# Display the missing values count
missing_values_count.show()
```

```
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Create a Spark session
spark = SparkSession.builder.appName("predictive-maintenance-analysis").getOrCreate()

# Count occurrences of 0 and 1 in the 'fault_detected' column
fault_detected_count = FaultDataset_DF.groupBy('fault_detected') \
                                     .count() \
                                     .withColumnRenamed('count', 'occurrences')

# Show the results
fault_detected_count.show()
```

```
import matplotlib.pyplot as plt

# Convert the result to Pandas DataFrame for visualization
fault_detected_count_pd = fault_detected_count.toPandas()

# Create a pie chart
plt.pie(fault_detected_count_pd['occurrences'], labels=fault_detected_count_pd['fault_detected'], autopct='%1.1f%%')
plt.title('Distribution of Fault/No Fault Detected')
plt.show()
```

```
#preprocess data into the correct format
## Import the RFormula class from the pyspark.ml.feature module
from pyspark.ml.feature import RFormula

## Define an instance of the RFormula class with the formula "fault_detected ~ .",
preprocess = RFormula(formula="fault_detected ~ .")

# Fit the RFormula transformation to the FaultDataset_DF DataFrame and apply RFormular transformation
FaultDataset_DF = preprocess.fit(FaultDataset_DF).transform(FaultDataset_DF)

#show first 5 rows
FaultDataset_DF.show(5)
```

```
# split data into training and test datasets

(trainingDF, testDF) = FaultDataset_DF.randomSplit([0.7, 0.3], seed=100)
```

```
from pyspark.ml.classification import DecisionTreeClassifier

dt = DecisionTreeClassifier(labelCol="label", featuresCol="features")

#train the model

model = dt.fit(trainingDF)
```

```
# make pedictions on the test dataset

predictions = model.transform(testDF)
predictions.show()
```

```
# use evaluator to measure accuracy of predictions of test data
# Use MulticlassClassificationEvaluator to evaluate the accuracy of predictions on the test data
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Set the label column name to "label", the prediction column name to "prediction",and the metric name to "accuracy"
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")

# Use the evaluator to calculate the accuracy of the predictions made on the test data
accuracy = evaluator.evaluate(predictions)

# Print out the accuracy of the predictions
print("Accuracy = %g " % (accuracy))
```

```
from pyspark.ml.tuning import ParamGridBuilder

#create a parameter grid

parameters = ParamGridBuilder()\
.addGrid(dt.impurity,["gini", "entropy"])\
.addGrid(dt.maxDepth, [4, 5, 6, 7, 8])\
.addGrid(dt.maxBins, [16, 24, 32, 44, 64])\
.build()
```

```
from pyspark.ml.tuning import TrainValidationSplit

#Define TrainValidationSplit

tvs = TrainValidationSplit()\
.setSeed(100)\
.setTrainRatio(0.75)\
.setEstimatorParamMaps(parameters)\
.setEstimator(dt)\
.setEvaluator(evaluator)
```

```
#train the model using grid search
gridsearchModel = tvs.fit(trainingDF)
```

```
#select best model and identify the hyperparameters
bestModel = gridsearchModel.bestModel

print("Parameters for the best Model:")
print("MaxDepth Parameter: %g" %bestModel.getMaxDepth())
print("Impurity Parameter: %s" %bestModel.getImpurity())
print("MaxBins Parameter: %g" %bestModel.getMaxBins())
```

```
#use the best model to make predictions on the hold out test set
evaluator.evaluate(bestModel.transform(testDF))
```

```
from pyspark.sql.functions import col

FaultDataset_DF = spark.read.csv('/FileStore/tables/FaultDataset.csv', header=True, inferSchema=True)

# Explore the data
FaultDataset_DF.show(5)
```

```
from pyspark.ml.feature import RFormula

# Preprocess the data into the correct format
preprocess = RFormula(formula="fault_detected ~ .")
FaultDataset_DF = preprocess.fit(FaultDataset_DF).transform(FaultDataset_DF)
FaultDataset_DF.show(5)
```

```
(trainingDF, testDF) = FaultDataset_DF.randomSplit([0.7, 0.3], seed=100)
```

```
from pyspark.ml.classification import LogisticRegression

# Train a logistic regression model
lr = LogisticRegression(labelCol="label", featuresCol="features")
lrModel = lr.fit(trainingDF)

# Use the model to make predictions on the test data
lrPredictions = lrModel.transform(testDF)

# Evaluate the accuracy of the model
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

lrEvaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
lrAccuracy = lrEvaluator.evaluate(lrPredictions)

print("Logistic Regression accuracy = %g " % (lrAccuracy))
```

```
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder

# Define the logistic regression model
lr = LogisticRegression(featuresCol='features', labelCol='label')

# Define the hyperparameter grid
paramGrid = ParamGridBuilder() \
    .addGrid(lr.regParam, [0.01, 0.1, 1]) \
    .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0]) \
    .addGrid(lr.maxIter, [10, 100, 1000]) \
    .build()

# Define the evaluator
evaluator = BinaryClassificationEvaluator(labelCol='label')

# Define the cross-validation object
cv = CrossValidator(estimator=lr, estimatorParamMaps=paramGrid, evaluator=evaluator)

# Train the model using cross-validation
cvModel = cv.fit(trainingDF)

# Choose the best model based on the evaluation metric
bestModel = cvModel.bestModel

# Evaluate the new model on the test set
predictions = bestModel.transform(testDF)
evaluator.evaluate(predictions)
```

```
# Load the CSV file as a Spark DataFrame with header and infer the schema
FaultDataset_DF = spark.read.csv('/FileStore/tables/FaultDataset.csv', header=True, inferSchema=True)

from pyspark.ml.feature import VectorAssembler

# Create a VectorAssembler to transform the input features into a single column
assembler = VectorAssembler(inputCols=FaultDataset_DF.columns[:-1], outputCol="features")

## Transform the DataFrame
data = assembler.transform(FaultDataset_DF)

# Split the data into training and test sets(70/30) with a set seed of 100
(trainingData, testData) = data.randomSplit([0.7, 0.3], seed=100)

from pyspark.ml.classification import RandomForestClassifier

# Define the RandomForestClassifier with a label column and feature column, and specify hyperparameters
rf = RandomForestClassifier(labelCol="fault_detected", featuresCol="features", numTrees=100, maxDepth=4)

# Fit the RandomForestClassifier to the training data
rfModel = rf.fit(trainingData)

# Make predictions on the testing data
predictions = rfModel.transform(testData)

from pyspark.ml.evaluation import MulticlassClassificationEvaluator

#Evaluate the accuracy of the predictions 
evaluator = MulticlassClassificationEvaluator(labelCol="fault_detected", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
```

```
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.classification import RandomForestClassifier

# Define the RandomForestClassifier model
rf = RandomForestClassifier(labelCol="fault_detected", featuresCol="features")

# Define the parameter grid to search over
paramGrid = (ParamGridBuilder()
             .addGrid(rf.numTrees, [10, 50, 100])
             .addGrid(rf.maxDepth, [5, 10, 20])
             .addGrid(rf.impurity, ["gini", "entropy"])
             .build())

# Define the evaluator to use for cross-validation
evaluator = MulticlassClassificationEvaluator(labelCol="fault_detected", predictionCol="prediction", metricName="accuracy")

# Define the cross-validation method and run the search
cv = CrossValidator(estimator=rf, estimatorParamMaps=paramGrid, evaluator=evaluator, numFolds=5)
cvModel = cv.fit(trainingData)

# Use the best model to make predictions on the test data
rfPredictions = cvModel.transform(testData)

# Evaluate the accuracy of the best model
rfAccuracy = evaluator.evaluate(rfPredictions)
print("Random Forest accuracy = %g" % rfAccuracy)
```
